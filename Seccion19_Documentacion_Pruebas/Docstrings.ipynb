```python
#documentar codigo en python 

class Pokemon:
    """Este es el docstring de la clase. 
    Esta es mi clase pokemon donde agrego un nuevo pokemon a la lista"""
    
    lista_pokedex=[]
    
    def __init__(self,nombre,edad):
        """Este es mi constructor de la clase recibe paramatros nombre y edad """
        self.nombre=nombre
        self.edad=edad
    
    def agregar(self):
        """este metodo agregar agrega un nuevo pokemon ala lista lista_pokedex """
        poke={"nombre":self.nombre,"edad":self.edad}
        self.lista_pokedex.append(poke)
    
    def __str__(self):
        """Este es un metodo especial que retorna una cadena de caracteres"""
        return "El pokemon {} con la edad: {}".format(self.nombre,self.edad) 
```

```python
objeto=Pokemon("Charizard",25)
objeto.agregar()
print(objeto)
```
